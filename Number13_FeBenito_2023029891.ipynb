```python
import time
import numpy as np
import matplotlib.pyplot as plt

def test_sorted_performance(sizes):
    times = []

    for size in sizes:
        data = np.random.randint(0, 10000, size).tolist()

        start_time = time.time()
        sorted_data = sorted(data)
        end_time = time.time()

        elapsed_time = end_time - start_time
        times.append(elapsed_time)

    return times

sizes = [100, 1000, 5000, 10000, 20000, 50000, 100000]

execution_times = test_sorted_performance(sizes)

plt.figure(figsize=(10, 6))
plt.plot(sizes, execution_times, marker='o')
plt.title("Performance of Python's sorted() Method")
plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.xscale("log")
plt.yscale("log")
plt.grid(True)
plt.show()
```
